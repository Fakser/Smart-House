```python
from K_MEAN import Model
import matplotlib.pyplot as plt
import random
import numpy as np
from sklearn.datasets import make_blobs, make_moons, make_checkerboard
import time
```

```python

X = [[np.random.normal() for _ in range(2)] for _ in range(3000)]
plt.scatter([x[0] for x in X], [x[1] for x in X])
plt.show()
```

```python
knn = Model(X)
start = time.time()
knn.fit(epochs = 100, verbose = 1)
time.time() - start
```

```python
fig = plt.figure()
plt.scatter([X[x][0] for x in knn.clusters[0]], [X[x][1] for x in knn.clusters[0]], c='r', s = 5)
plt.scatter(knn.centroids[0][0], knn.centroids[0][1], c='y', s = 120)
plt.scatter([X[x][0] for x in knn.clusters[1]], [X[x][1] for x in knn.clusters[1]], c='g', s = 5)
plt.scatter(knn.centroids[1][0], knn.centroids[1][1], c='y', s = 120)
plt.scatter([X[x][0] for x in knn.clusters[2]], [X[x][1] for x in knn.clusters[2]], c='b', s = 5)
plt.scatter(knn.centroids[2][0], knn.centroids[2][1], c='y', s = 120)

plt.show()
```
